```python
class BST:
    def __init__(self, value, left = None, right = None):
        self.value = value
        self.left = left
        self.right = right
    
    def insert(self, value):
        if self.value == value:
            return
        elif self.value < value:
            if self.right is None:
                self.right = BST(value)
            else:
                self.right.insert(value)
        else:
            if self.left is None:
                self.left = BST(value)
            else:
                self.left.insert(value)
    
    def find(self, value):
        if self.value == value:
            return True
        elif self.value < value:
            if self.right is None:
                return False
            else:
                self.right.find(value)
        else:
            if self.left is None:
                return False
            else:
                self.left.find(value)
    
    def _getMinValue(self):
        if self.left is None:
            return self.value
        else:
            return self.left.getMinValue()
        
    def remove(self, value):
        if self.value == value:
            if self.left is None and self.right is None:
                return None
            elif self.left is None:
                return self.right
            elif self.right is None:
                return self.left
            else:
                self.value = self.right._getMinValue()
                self.right = self.right.remove(self.value)
        elif self.value < value:
            self.right = self.right.remove(value)
        else:
            self.left = self.left.remove(value)
        return self
    
    def find(self, value):
        if self.value == value:
            return value
        elif self.left is not None:
            self.left.find(value)
        elif self.right is not None:
            self.right.find(value)
        else:
            return None
    
    def display(self):
        lines, *_ = self._display_aux()
        for line in lines:
            print(line)

    def _display_aux(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.value
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left._display_aux()#you can replace with w,h,m
            s = '%s' % self.value
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right._display_aux()#you can replace with w,h,m
            s = '%s' % self.value
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = self.left._display_aux()# (n =1,p=1,x=0) you can replace with l_w,l_h,l_m
        right, m, q, y = self.right._display_aux() #m=1,q=1,y=0 you can replace with r_w,r_h,r_m
        s = '%s' % self.value
        u = len(s)#u=1
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2
```

```python
import random

my_BST = BST(50)

for _ in range (10):
    my_BST.insert(random.randint(0,100))
    
print(my_BST.display())
```
